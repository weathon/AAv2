```python
import json
with open("commits.json.bak", "r") as f:
    data = json.load(f)
```

```python
methods = []
with open("../classes.json", "r") as f:
    classes = json.load(f)

for main_type in ["anti_aesthetics"]:
    print(f"Registered class: {main_type}")
    for method in classes[main_type]:
        print(method)
        methods.append(method)  
```

```python
ds = []
for i, commit in enumerate(data.keys()):
    subset = data[commit]["entries"]
    for j in range(len(subset)):
        subset[j]["commit"] = commit
    ds.extend(subset)
```
